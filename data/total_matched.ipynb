```python
import pandas as pd
```

```python
pd0 = pd.read_csv("MatchedResult_0.txt",header=None)
```

```python
def remove_punc(str_item):
    str_item = str_item.replace(')','')
    str_item = str_item.replace('(','')
    return str_item
```

```python
pd0 = pd.read_csv("MatchedResult_0.txt",header=None)

pd0['short'] = pd0.iloc[:, 0].apply(remove_punc)
pd0['long'] = pd0.iloc[:, 1].apply(remove_punc)

pd0 = pd0.iloc[:, 2:4]
```

```python
pd1 = pd.read_csv("MatchedResult_1.txt",header=None)

pd1['short'] = pd1.iloc[:, 0].apply(remove_punc)
pd1['long'] = pd1.iloc[:, 1].apply(remove_punc)

pd1 = pd1.iloc[:, 2:4]
```

```python
pd2 = pd.read_csv("MatchedResult.txt",header=None)

pd2['short'] = pd2.iloc[:, 0].apply(remove_punc)
pd2['long'] = pd2.iloc[:, 1].apply(remove_punc)

pd2 = pd2.iloc[:, 2:4]
```

```python
frames = [pd0, pd1, pd2]

pd_input1 = pd.concat(frames)
```
